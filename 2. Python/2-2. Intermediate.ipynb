### 2.2.3 Generator & Iterator

> **`list`** Ï†ÑÏ≤¥Î•º ÏÇ¨Ïö©ÌïòÎäî Í≤ÉÏùÄ ÎπÑÌö®Ïú®Ï†ÅÏù¥Îã§.

```python
def lazy_range(n):
    """rangeÏôÄ Í∞ôÏùÄ Í∏∞Îä•ÏùÑ ÌïòÎäî generator"""
    i = 0
    while i < n:
        yield i
        i += 1
```

```python
for i in lazy_range(10):
    print(i)
```

### 2.2.7 OOP

```python
class Set:
    
    """
    Î™®Îì† Î©§Î≤ÑÌï®ÏàòÏùò Ï≤´Î≤àÏß∏ params Îäî 'self' Îã§ (Í¥ÄÏäµ))
    'self'ÎûÄ ÌòÑÏû¨ ÏÇ¨Ïö©ÎêòÍ≥† ÏûàÎäî Set Í∞ùÏ≤¥Î•º ÏùòÎØ∏
    """
    
    def __init__(self, values=None):
        """
        Ïù¥Í±¥ ÏÉùÏÑ±Ïûê(Constructor)
        ÏÉàÎ°úÏö¥ SetÏùÑ ÎßåÎì§Î©¥ Ìò∏Ï∂úÎêúÎã§.
        """
        self.dict = {} # Ïù¥ dictÎ•º ÌÜµÌï¥ Ìï≠Î™©Ïùò Ï°¥Ïû¨ Ïó¨Î∂ÄÎ•º ÌôïÏù∏Ìï®
        
        if values is not None:
            for value in values:
                self.add(value)
                
    def __repr__(self):
        """
        Ïù¥ Ìï®ÏàòÎ•º ÏûÖÎ†•ÌïòÍ±∞ÎÇò str()Î°ú Î≥¥ÎÇ¥Ï£ºÎ©¥
        Set Í∞ùÏ≤¥Î•º Î¨∏ÏûêÏó¥Î°ú ÌëúÌòÑÌï¥ Ï§å
        """
        return "Set: " + str(self.dict.keys())
    
    
    # self.dictÏóêÏÑú Ìï≠Î™©Í≥º TrueÎ•º Í∞ÅÍ∞Å keyÏôÄ valueÎ°ú ÏÇ¨Ïö©Ìï¥ÏÑú
    # Set ÏïàÏóê Ï°¥Ïû¨ÌïòÎäî Ìï≠Î™©ÏùÑ ÌëúÌòÑ
    
    def add(self, value):
        self.dict[value] = True
        
    # ÎßåÏïΩ Ìï≠Î™©Ïù¥ dictÏùò key ÎùºÎ©¥ Ìï≠Î™©ÏùÄ Set ÏïàÏóê Ï°¥Ïû¨Ìï®
    def contains(self, value):
        return value in self.dict
    
    def remove(self, value):
        del self.dict[value]
```

```python
s = Set([1, 2, 3])
s.add(4)
print(s.contains(4))
s.remove(3)
print(s.contains(3))
s.__repr__()
```

### map, reduce, filter

```python
def double(x):
    return 2*x
```

```python
xs = [1, 2, 3, 4]
twice_xs = [double(x) for x in xs]
print(twice_xs)
twice_xs2 = list(map(double, xs))
print(twice_xs2)
```

### 2.2.8 Enumerate

> `list`Ïùò Ìï≠Î™©Í≥º index Î™®Îëê ÌïÑÏöîÌïú Í≤ΩÏö∞

```python
# not pythonic
for i in range(len(documents)):
    document = documents[i]
    do_something(i, document)
    

# more pythonic but not yet
i = 0
for document in documents:
    do_something(i, document)
    i += 1
```

```python
# most pythonic üòÄ

for i, document in enumerate(documents):
    do_something(i, document)
```

> `index`Îßå ÌïÑÏöîÌïú Í≤ΩÏö∞

```python
for i in range(len(documents)): do_something(i)  # not pythonic
for i, _ in enumerate(documents): do_something(i) # most pythonic üòÄ
```
